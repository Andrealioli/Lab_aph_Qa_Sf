# Índice do efeito de empacotamento

Para estimar o índice do efeito de empacotamento da Bricaud et al. (2004):
$${Q_{a}^*} = a_{ph} (\lambda) + a_{sol} (\lambda) $$
Mas antes é necessário estimar o :
$$ a_{sol} (\lambda) = \sum C_{i}.a_{sol, i}^*(\lambda) $$
Gráfico dos $a_{sol, i}^*(\lambda)$ de cada pigmento dado pelo HPLC:

```python
#Carregando pacotes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

bricaud_asol = pd.read_csv("Bricaud_et_al_2004.csv", skiprows=4, na_values=999 )
bricaud_asol.rename(columns = {'Chla ':'Chla' ,"Dv-Chla": "DVChla", "Chl b": "Chlb", 
                               "Dv-Chlb":"DVChlb", "FUCO":"Fuco", "19'-BF":"Butfuco", 
                               "19'-HF":"Hexfuco", "beta-Car":"betacar", "a-car":"acar"}, inplace = True) 
```

```python
bricaud_asol.head()
```

```python
df=bricaud_asol.cumsum()
plt.plot(bricaud_asol['lambda'],bricaud_asol[['Chla','DVChla','Chlb', 'DVChlb','Chlc12',
 'Fuco','Butfuco','Hexfuco','Peri','Diad','Zea','Allox','betacar','acar']])
plt.xlabel("Wavelength (nm)")
plt.ylabel(r"$a_{sol}^* (m^2 mg^{-1})$")

plt.show()

```
